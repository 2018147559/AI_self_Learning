필수과제 : 

```python
import requests
from bs4 import BeautifulSoup
```

```python
headers= {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"}
```

```python
for i in range(1,51) :#50페이지까지 있어서!
    url = 'https://comic.naver.com/webtoon/list?titleId=183559&weekday=mon&page={}'.format(i)
    res = requests.get(url,headers = headers)
    soup = BeautifulSoup(res.text, "lxml")
    ul=soup.find_all("td",attrs={"class":"title"})
    for td in ul:
        title = td.find("a").get_text().strip()
        link = "https://comic.naver.com" + td.find("a")['href']
        print("소제목  " + title)
        print("(링크 : " + link + ")")
```
